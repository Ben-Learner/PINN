```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import scipy.io 

#加载数据
data_path = './Project_IV_Data.mat'
data = scipy.io.loadmat(data_path)

# 数据分割
u_bc = data['u_bc']
u_data = data['u_data']
v_data = data['v_data']
x_2d = data['x_2d']
y_2d = data['y_2d']

# 将numpy数组转换为torch张量
u_bc_tensor = torch.tensor(u_bc, dtype=torch.float32)
u_data_tensor = torch.tensor(u_data, dtype=torch.float32)
v_data_tensor = torch.tensor(v_data, dtype=torch.float32)
x_2d_tensor = torch.tensor(x_2d.flatten(), dtype=torch.float32)
y_2d_tensor = torch.tensor(y_2d.flatten(), dtype=torch.float32)

# 创建网格点张量
grid_tensor = torch.stack((x_2d_tensor, y_2d_tensor), dim=1)

# 分割训练和测试数据
train_bc = u_bc_tensor[:90]
test_bc = u_bc_tensor[90:]
train_u_data = u_data_tensor[:90]
test_u_data = u_data_tensor[90:]
train_v_data = v_data_tensor[:90]
test_v_data = v_data_tensor[90:]

# 扩展边界条件数据以匹配速度场的形状
def expand_boundary_conditions(bc_data, grid_size):
    """扩展边界条件数据到完整的网格尺寸。"""
    expanded_bc = torch.zeros((bc_data.shape[0], grid_size, grid_size))
    expanded_bc[:, 0, :] = bc_data  # 上边界具有非零值
    return expanded_bc

grid_size = 65  # 网格大小
expanded_train_u_bc = expand_boundary_conditions(train_bc, grid_size)
expanded_test_u_bc = expand_boundary_conditions(test_bc, grid_size)

# 创建PyTorch数据集和加载器
train_dataset = TensorDataset(expanded_train_u_bc, train_u_data, train_v_data)
test_dataset = TensorDataset(expanded_test_u_bc, test_u_data, test_v_data)

train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# 确认扩展后的边界条件数据集维度
print(expanded_train_u_bc.shape, expanded_test_u_bc.shape)

```

```python
# 定义PINN模型
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        # 增加输入维度以包含扩展的边界条件
        self.fc1 = nn.Linear(3, 50)  # 输入维度为3（x坐标、y坐标和边界条件）
        self.fc2 = nn.Linear(50, 100)
        self.fc3 = nn.Linear(100, 100)
        self.fc4 = nn.Linear(100, 2)  # 输出2维，速度u和v

        self.activation = nn.Tanh()
        
    def forward(self, x, bc):
        # 扁平化边界条件和网格张量，以便可以合并
        flat_bc = bc.view(-1, 1)
        flat_grid = x.repeat(bc.size(0), 1, 1).view(-1, 2)
        
        # 合并网格张量和边界条件
        combined_input = torch.cat((flat_grid, flat_bc), dim=1)

        x = self.activation(self.fc1(combined_input))
        x = self.activation(self.fc2(x))
        x = self.activation(self.fc3(x))
        x = self.fc4(x)
        return x


# 实例化模型和优化器
pinn = PINN()
optimizer = optim.Adam(pinn.parameters(), lr=1e-4)

# 训练函数
def train(model, train_loader, optimizer, epochs, grid_tensor):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for bc, u_data, v_data in train_loader:
            optimizer.zero_grad()
            
            # 预测整个域的速度场（考虑边界条件）
            u_pred = model(grid_tensor, bc).reshape(-1, 65, 65, 2)
            # 调整目标数据的形状以匹配预测的形状
            u_data = u_data.squeeze(0)  # 移除批次维度
            v_data = v_data.squeeze(0)  # 移除批次维度
            # 计算损失函数
            loss_u = nn.MSELoss()(u_pred[:, :, :, 0], u_data)
            loss_v = nn.MSELoss()(u_pred[:, :, :, 1], v_data)

            # 总损失
            loss = loss_u + loss_v
            
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        
        if (epoch+1) % 10 == 0:
            print(f'Epoch {epoch+1}, Loss: {total_loss / len(train_loader)}')

# 训练模型
train(pinn, train_loader, optimizer, epochs=100, grid_tensor=grid_tensor)

# 测试函数
def test(model, test_loader, grid_tensor):
    model.eval()  # 设置模型为评估模式
    total_loss = 0
    with torch.no_grad():  # 关闭梯度计算
        for bc, u_data, v_data in test_loader:
            # 预测整个域的速度场
            u_pred = model(grid_tensor, bc).reshape(-1, 65, 65, 2)
            
            # 计算损失
            # 调整目标数据的形状以匹配预测的形状
            u_data = u_data.squeeze(0)  # 移除批次维度
            v_data = v_data.squeeze(0)  # 移除批次维度
            loss_u = nn.MSELoss()(u_pred[:, :, :, 0], u_data)
            loss_v = nn.MSELoss()(u_pred[:, :, :, 1], v_data)
            loss = loss_u + loss_v
            
            total_loss += loss.item()
    
    print(f'Test Loss: {total_loss / len(test_loader)}')

# 测试模型
test(pinn, test_loader, grid_tensor)
```

```python
# 可视化函数
import matplotlib.pyplot as plt
def visualize_results(model, grid_tensor, expanded_u_bc):
    model.eval()  # 设置模型为评估模式
    with torch.no_grad():  # 关闭梯度计算
        # 预测速度场
        # 由于模型现在需要边界条件作为输入，我们将使用一组扩展的边界条件进行预测
        example_bc = expanded_u_bc[8].unsqueeze(0)  # 使用第一个案例的边界条件进行演示
        u_pred = model(grid_tensor, example_bc).reshape(65, 65, 2)
        u_pred = u_pred.detach().numpy()

    # 绘制速度场的u分量
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(u_pred[:, :, 0], origin='lower', extent=[0, 1, 0, 1])
    plt.colorbar()
    plt.title('Predicted u-component of Velocity')

    # 绘制速度场的v分量
    plt.subplot(1, 2, 2)
    plt.imshow(u_pred[:, :, 1], origin='lower', extent=[0, 1, 0, 1])
    plt.colorbar()
    plt.title('Predicted v-component of Velocity')

    plt.show()

# 可视化训练后的模型结果
visualize_results(pinn, grid_tensor, expanded_test_u_bc)
```
